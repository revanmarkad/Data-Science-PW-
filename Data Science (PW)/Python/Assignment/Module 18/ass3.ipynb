Q 4 Explain the following with python code :read(),readline() and readlines().

In Python, the `read()`, `readline()`, and `readlines()` methods are used to read data from a file. Each method has different behavior and use cases. Here's an explanation of each, along with examples:

### 1. `read()`

The `read()` method reads the entire contents of a file and returns it as a string. It can also accept an optional parameter specifying the number of bytes to read.

#### Example:

```python
# Create a file with some content
with open("example.txt", "w") as file:
    file.write("Hello, World!\nWelcome to Python file handling.\nHave a great day!")

# Read the entire file content
with open("example.txt", "r") as file:
    content = file.read()
    print("Using read():")
    print(content)
```

#### Output:
```
Using read():
Hello, World!
Welcome to Python file handling.
Have a great day!
```

### 2. `readline()`

The `readline()` method reads a single line from the file. It moves the file pointer to the start of the next line after reading.

#### Example:

```python
# Read one line at a time
with open("example.txt", "r") as file:
    print("Using readline():")
    first_line = file.readline()
    print("First Line:", first_line)
    
    second_line = file.readline()
    print("Second Line:", second_line)
```

#### Output:
```
Using readline():
First Line: Hello, World!

Second Line: Welcome to Python file handling.
```

### 3. `readlines()`

The `readlines()` method reads all the lines of the file and returns them as a list of strings, where each string is a line from the file.

#### Example:

```python
# Read all lines as a list
with open("example.txt", "r") as file:
    lines = file.readlines()
    print("Using readlines():")
    for line in lines:
        print(line, end="")  # Use end="" to avoid adding extra newlines
```

#### Output:
```
Using readlines():
Hello, World!
Welcome to Python file handling.
Have a great day!
```

### Summary:

- **`read()`**: Reads the entire file or a specified number of bytes and returns it as a single string.
- **`readline()`**: Reads one line at a time and returns it as a string.
- **`readlines()`**: Reads all lines and returns them as a list of strings.

Each method is useful depending on how you want to process the file data. Use `read()` for simple cases where you need all the data at once, `readline()` when you need to process lines one by one, and `readlines()` when you need to work with all lines as a list.
