Q 4 Using the database and the collection created in question number 3, write a code to insert one rrecord , and insert many records.  Use the find() and find_one() method to print the inserted record.

 

```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Access the Database and Collection
db = client["ecommerce_db"]
customers_collection = db["customers"]

# Insert One Record
single_customer = {
    "first_name": "Bob",
    "last_name": "Brown",
    "email": "bob.brown@example.com",
    "address": {
        "street": "456 Oak Street",
        "city": "Wonderland",
        "zip": "12345"
    },
    "orders": [
        {
            "order_id": "B001",
            "product": "Tablet",
            "amount": 500.00,
            "date": "2023-03-10"
        }
    ]
}
insert_one_result = customers_collection.insert_one(single_customer)

# Print the inserted record using find_one()
print("Inserted One Record:")
print(customers_collection.find_one({"_id": insert_one_result.inserted_id}))

# Insert Many Records
many_customers = [
    {
        "first_name": "Charlie",
        "last_name": "Clark",
        "email": "charlie.clark@example.com",
        "address": {
            "street": "789 Pine Street",
            "city": "Wonderland",
            "zip": "12345"
        },
        "orders": [
            {
                "order_id": "C001",
                "product": "Monitor",
                "amount": 300.00,
                "date": "2023-04-05"
            }
        ]
    },
    {
        "first_name": "Diana",
        "last_name": "Davis",
        "email": "diana.davis@example.com",
        "address": {
            "street": "101 Birch Street",
            "city": "Wonderland",
            "zip": "12345"
        },
        "orders": [
            {
                "order_id": "D001",
                "product": "Keyboard",
                "amount": 100.00,
                "date": "2023-05-01"
            }
        ]
    }
]
insert_many_result = customers_collection.insert_many(many_customers)

# Print the inserted records using find()
print("\nInserted Many Records:")
for customer_id in insert_many_result.inserted_ids:
    print(customers_collection.find_one({"_id": customer_id}))

# Retrieve and Print All Documents in the Collection
print("\nAll Records in Collection:")
for doc in customers_collection.find():
    print(doc)
```

 

1. **Insert One Record**: Inserts a single customer document into the `customers` collection.
2. **Insert Many Records**: Inserts multiple customer documents into the `customers` collection.
3. **find_one() Method**: Retrieves and prints the single inserted document using its `_id`.
4. **find() Method**: Retrieves and prints all inserted documents using a loop.
5. **Print All Documents**: Retrieves and prints all documents currently in the `customers` collection.
